```python
data = pd.read_csv("HRDataset_v14.csv")
data.head()
```

```python
#select desired columns
data = data[["RecruitmentSource",	"PerformanceScore",	"EngagementSurvey",	"EmpSatisfaction",	"SpecialProjectsCount",	"DaysLateLast30"]]
```

```python
data_encoded = pd.get_dummies(data, prefix_sep="_")
data_encoded = data_encoded.dropna()
data_encoded.head()
```

```python
data_encoded.head()
```

```python
x = data_encoded.drop(columns="PerformanceScore_PIP")
y = data_encoded["PerformanceScore_PIP"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.2)
print(Counter(y_train))
```

#SVM

```python
clf = svm.SVC(C=1, kernel="rbf")
clf.fit(X_train,y_train)
```

```python
y_predict = clf.predict(X_test)
print("Accuracy:",accuracy_score(y_test,y_predict))

print(classification_report(y_test,y_predict))
```

#GradientBoostingClassier

```python
cls =GradientBoostingClassifier(n_estimators=50,learning_rate=0.01)
cls.fit(X_train,y_train)
```

```python
y_predict = cls.predict(X_test)
print("Accuracy:",accuracy_score(y_test,y_predict))
print(classification_report(y_test,y_predict))
```
